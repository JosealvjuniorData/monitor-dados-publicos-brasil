```python
import streamlit as st
import requests
import pandas as pd

st.set_page_config(layout="wide", page_title="Navegador de Dados Abertos")

st.title("üèõÔ∏è Navegador de Dados Abertos do Brasil")

# 1. Identifica√ß√£o para evitar bloqueio do governo
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

# --- BUSCA POR TEMA ---
tema_input = st.sidebar.text_input("1. Digite um tema (ex: Saude, Educacao):", "Saude")

if tema_input:
    # A URL precisa de par√¢metros de busca limpos
    url_busca = f"https://dados.gov.br/api/3/action/package_search?q={tema_input}&rows=10"
    
    try:
        # Fazemos a requisi√ß√£o com os headers de navegador real
        response = requests.get(url_busca, headers=headers, timeout=15)
        
        # S√≥ tentamos ler o JSON se a resposta for 200 (OK)
        if response.status_code == 200:
            res_json = response.json()
            
            if res_json.get('success') and res_json['result']['results']:
                datasets = {ds['title']: ds['id'] for ds in res_json['result']['results']}
                
                # --- SELE√á√ÉO DO CONJUNTO ---
                dataset_nome = st.sidebar.selectbox("2. Selecione o Conjunto de Dados:", list(datasets.keys()))
                st.write(f"Voc√™ selecionou: **{dataset_nome}**")
                st.info("Agora voc√™ pode expandir para listar as tabelas deste conjunto!")
            else:
                st.warning("Nenhum dado encontrado para este tema no momento.")
        else:
            st.error(f"O portal do governo recusou a conex√£o (Erro {response.status_code}).")
            st.info("Dica: Verifique se voc√™ consegue acessar o site dados.gov.br manualmente no seu navegador.")

    except Exception as e:
        st.error(f"Erro inesperado: {e}")
```
